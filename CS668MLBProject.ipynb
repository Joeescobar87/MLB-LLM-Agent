```python
def result_to_text(result):
    if isinstance(result, pd.DataFrame):
        if result.empty:
            return "No rows returned."
        return result.to_string(index=False)

    if isinstance(result, dict):
        return json.dumps(result, indent=2)

    return str(result)
```

```python
def run_tool(tool_name, args):

    if tool_name == "getHitterLeaderboard":
        return baseball_tools.getHitterLeaderboard(
            df_statcast_clean,
            df_hitters,
            metric=args.get("metric", "HR"),
            min_ab=args.get("min_ab", 100),
            start_date=args.get("start_date"),
            end_date=args.get("end_date"),
            top_n=args.get("top_n", 10),
            ascending=args.get("ascending", False)
        )

    if tool_name == "getPitcherLeaderboard":
        return baseball_tools.getPitcherLeaderboard(
            df_statcast_clean,
            df_pitchers,
            metric=args.get("metric", "SO"),
            min_bf=args.get("min_bf", 100),
            start_date=args.get("start_date"),
            end_date=args.get("end_date"),
            top_n=args.get("top_n", 10),
            ascending=args.get("ascending", False)
        )

    if tool_name == "compareHitters":
        return baseball_tools.compareHitters(
            df_statcast_clean,
            df_hitters,
            args["player1"],
            args["player2"],
            start_date=args.get("start_date"),
            end_date=args.get("end_date")
        )

    if tool_name == "comparePitchers":
        return baseball_tools.comparePitchers(
            df_statcast_clean,
            df_pitchers,
            args["player1"],
            args["player2"],
            start_date=args.get("start_date"),
            end_date=args.get("end_date")
        )

    if tool_name == "getHitterLeaderboardVsPitcherHand":
        return baseball_tools.getHitterLeaderboardVsPitcherHand(
            df_statcast_clean,
            df_hitters,
            pitcher_hand=args.get("pitcher_hand", "L"),
            metric=args.get("metric", "BA"),
            min_ab=args.get("min_ab", 50),
            top_n=args.get("top_n", 10),
            ascending=args.get("ascending", False),
            start_date=args.get("start_date"),
            end_date=args.get("end_date")
        )

    if tool_name == "hitterVsPitcher":
        return baseball_tools.hitterVsPitcher(
            df_statcast_clean,
            df_hitters,
            args["hitter_name"],
            args["pitcher_name"]
        )

    if tool_name == "getHitterStatsVsPitchType":
        allowed_metrics = [
            "PA", "AB", "H", "HR", "TB", "BB", "SO",
            "BA", "SLG", "K_rate", "BB_rate"
        ]

        metric = args.get("metric", "BA")

        if metric not in allowed_metrics:
            metric = "BA"

        return baseball_tools.getHitterStatsVsPitchType(
            df_statcast_clean,
            df_hitters,
            pitch_name=args["pitch_name"],
            hitter_name=args.get("hitter_name"),
            metric=metric,
            min_ab=args.get("min_ab", 20),
            top_n=args.get("top_n"),
            descending=args.get("descending", False)
        )

    if tool_name == "getSingleHitterStats":
        return baseball_tools.getSingleHitterStats(
            df_statcast_clean,
            df_hitters,
            hitter_name=args["hitter_name"],
            start_date=args.get("start_date"),
            end_date=args.get("end_date")
        )
    if tool_name == "getSinglePitcherStats":
        return baseball_tools.getSinglePitcherStats(
            df_statcast_clean,
            df_pitchers,
            pitcher_name=args["pitcher_name"],
            start_date=args.get("start_date"),
            end_date=args.get("end_date")
        )

    return f"Unsupported tool: {tool_name}"
```

```python
def ask_agent(question, show_debug=False):
    response = client.responses.create(
    model=model,
        input=[
            {
                "role": "system",
                "content": """
You are an MLB analytics agent.

Use the provided tools for MLB stat questions.
Do not answer from memory.
The dataset is the 2025 MLB regular season.

Date rules:
- If the user asks about May, use start_date='2025-05-01' and end_date='2025-05-31'.
- If the user asks about June, use start_date='2025-06-01' and end_date='2025-06-30'.
- If the user says through July 15, use end_date='2025-07-15'.
- If the user says before July 15, use end_date='2025-07-14'.

Metric rules:
- home runs -> HR
- batting average -> BA
- slugging -> SLG
- hits -> H
- pitcher strikeouts -> getPitcherLeaderboard metric SO
- hitter strikeouts -> getHitterLeaderboard metric SO
- ERA -> getPitcherLeaderboard metric ERA_official and ascending true
- WHIP -> getPitcherLeaderboard metric WHIP_official and ascending true
- lowest/fewest/least -> ascending true
- highest/most/best -> ascending false

Pitcher hand:
- left-handed pitchers or lefties -> pitcher_hand L
- right-handed pitchers or righties -> pitcher_hand R

Pitch type:
- sliders -> pitch_name Slider
- fastballs -> pitch_name 4-Seam Fastball unless the user says another fastball type
- curveballs -> pitch_name Curveball
- changeups -> pitch_name Changeup

Default filters:
- min_ab=100 for full-season hitter leaderboards.
- min_ab=20 for monthly or pitch-type hitter leaderboards.
- min_ab=50 for pitcher-hand split hitter leaderboards.
- min_bf=100 for full-season pitcher leaderboards.
- min_bf=20 for monthly pitcher leaderboards.
- top_n=10 unless the user asks for a specific number.
"""
            },
            {
                "role": "user",
                "content": question
            }
        ],
        tools=agent_tools,
        tool_choice="required"
    )

    tool_call = None

    for item in response.output:
        if item.type == "function_call":
            tool_call = item
            break

    if tool_call is None:
        return response.output_text

    tool_name = tool_call.name
    args = json.loads(tool_call.arguments)

    tool_result = run_tool(tool_name, args)

    final_response = client.responses.create(
    model="gpt-5",
        input=f"""
User question:
{question}

Tool used:
{tool_name}

Tool arguments:
{json.dumps(args, indent=2)}

Tool result:
{result_to_text(tool_result)}

Answer the user's question clearly using only the tool result.
Do not invent stats.
If the sample size is small, mention that.
Do not mention tool arguments unless useful.
"""
    )

    if show_debug:
        return {
            "question": question,
            "tool_name": tool_name,
            "arguments": args,
            "tool_result": tool_result,
            "answer": final_response.output_text
        }

    return final_response.output_text
```
